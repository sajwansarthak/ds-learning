## Data Transformation

```python
df_aggbookings
```

```python
df_aggbookings["occ_pst"] = df_aggbookings["successful_bookings"]/df_aggbookings["capacity"]
df_aggbookings.head(4)
```

```python
df_aggbookings["occ_pst"] = df_aggbookings["occ_pst"].apply(lambda x : round(x*100,2))
df_aggbookings.head(4)
```

## 1-> What is the avg occupancy rate in each of the room category

```python
df_aggbookings.groupby("room_category")["occ_pst"].mean().round(2)
```

```python
df = pd.merge(df_aggbookings,df_rooms,left_on = "room_category",right_on = "room_id")
df.head(4)
```

```python
df.drop("room_id",axis=1,inplace=True)
df.head(4)
```

```python
df.groupby("room_class")["occ_pst"].mean().round(2)
```

## 2->Avg occupancy rate per city

```python
df = pd.merge(df,df_hotels, on="property_id")
df.head(4)
```

```python
df.groupby("city")["occ_pst"].mean().round(2)
```

## 3->When was the occupancy better ? weekdays or weekends ?

```python
df = pd.merge(df,df_date,left_on="check_in_date",right_on="date")
df.head(4)
```

```python
df.groupby("day_type")["occ_pst"].mean().round(2)
```

## 4->In the month of june what is the occupancy for different cities

```python
df["mmm yy"].unique()
```

```python
df_june_22 = df[df["mmm yy"] == "Jun 22"]
df_june_22.head(3)
```

```python
df_june_22.groupby("city")["occ_pst"].mean().round(2).sort_values(ascending=False)
```

```python
new_df = pd.concat([df,df_aug],ignore_index=True,axis=0)
new_df.tail(10)
```
