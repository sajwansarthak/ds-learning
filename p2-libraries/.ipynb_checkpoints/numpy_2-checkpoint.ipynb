```python
import numpy as np
```

```python
#Sales for quarter 1 (Matrix 1)
#Rows represent different products and columns represent different regions 


q1 = np.array([
    [200,220,250], #product A
    [150,180,210], #product B
    [300,330,360]  #product C
])


q2 = np.array([
    [209,231,251], #product A
    [155,192,222], #product B
    [310,340,375]  #product C
])
```

```python
q1 + q2
```

```python
q2 - q1
```

```python
#percentage increase
(q2 - q1)*100/q1
```

```python
prices = np.array([
    [10,12,11],  #prices of product A in different region
    [8,9,10],    #prices of b
    [15,16,17]   #prices of c
])

q1_revenue = q1 * prices
q1_revenue
```

```python
q1_dis = q1_revenue*0.2
q1_dis
```

```python
q1_net_rev = q1_revenue - q1_dis
q1_net_rev
```

```python
np.sum(q1_net_rev)
```
