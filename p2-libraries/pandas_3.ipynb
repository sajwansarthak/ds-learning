## Reading excel files and working on them 

```python
df_movies = pd.read_excel("movies_db.xlsx","movies")
df_movies.head(4)
```

```python
df_finan = pd.read_excel("movies_db.xlsx","financials")
df_finan.head(4)
```

```python
def stand_currency(curr):
    if curr =="$$" or curr == "Dollars":
       return "USD"
    return curr


df_finan = pd.read_excel("movies_db.xlsx","financials",converters ={
    'currency':stand_currency
})
df_finan.head(5)
```

```python
df_merged = pd.merge(df_movies,df_finan,on="movie_id")
df_merged
```

```python
df_merged.to_excel("movies_merged.xlsx",sheet_name="Merged",index = False)
```
